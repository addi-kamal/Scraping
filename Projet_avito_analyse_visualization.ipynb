# 4-Concatebation des dataframes

conctenation des dataframes est simple puisqu'ona le meme nombre des colonnes et elle sont du meme type 

```python
Annonces = pd.concat([Annonces_offres, Annonces_demandes], ignore_index = True )
```

```python
Annonces.head()
```

# 5-Informations sur les prix 

Afficher les produits dont le prix existe

```python
Annonces_prix_existe = Annonces_offres[Annonces_offres['Prix'].notnull()]
```

```python
Annonces_prix_existe.head()
```

les produits dont le prix est inferieur a 1000DH

```python
Annonces_prix_infer_a_1000 = Annonces[Annonces["Prix"] <= 1000]
```

```python
Annonces_prix_infer_a_1000.head()
```

sommes d'argents en DH par jour

```python
Somme_grouper_par_dates = Annonces.groupby("Dates")["Prix"].sum() 
```

```python
Somme_grouper_par_dates 
```

la moyenne par jour

```python
moy = Somme_grouper_par_dates.mean()
print("la moyenne est %.2f DH"%(moy))
```

la somme maximale et leur date

```python
somme_max = Somme_grouper_par_dates.max()
date_somme_max = Somme_grouper_par_dates.idxmax()    #l'indice de la valeur maximale dans la serie des prix par jour
print("la somme maximale des prix des annonces par jour est :", somme_max )
print("Enregistrer le ", date_somme_max)
```

la somme minimale et leur date

```python
somme_min = Somme_grouper_par_dates.min()
date_somme_min = Somme_grouper_par_dates.idxmin()    #l'indice de la valeur minimale 
print("la somme minimale des prix d'annonces par jour est :", somme_min )
print("Enregistrer le ", date_somme_min)
```

**variation des prix d'annonces en fonction des dates**

```python
indices = Somme_grouper_par_dates.index    
```

```python
plt.figure(figsize = (30,10))
plt.plot(indices, (Somme_grouper_par_dates[:])/10000)  # on devise par 10000DH pour que le graphique soit visible et facile a analyser
plt.xticks(rotation=70, fontsize=11)
plt.yticks( fontsize=14)
plt.xlabel("Dates ",fontsize = 20)
plt.ylabel("Prix",fontsize = 18)
plt.title("Variation des prix (en x10000DH) en fonction des dates", fontsize = 25)
plt.savefig('prix_en_fonction_dates.png', dpi=400, bbox_inches='tight')
```

# 6-Catégories

le nombre d'annonces de chaque catégories

```python
Serie_categorie = Annonces.groupby("Categories")["Categories"].count().sort_values(ascending=True)  #croissante par rapport au nombre de chaque categorie
```

```python
Serie_categorie.head(10)
```

les  10 categories les plus frequents :

```python
Serie_categorie[-10:]
```

```python
plt.figure(figsize=(15, 10))
plt.tight_layout()
plt.barh((Serie_categorie[-10:].index),Serie_categorie[-10:],color='blue')
plt.xticks( fontsize=15)
plt.yticks( fontsize=15)
plt.xlabel("Nombre de categories ",fontsize=15)
plt.ylabel("les 10 top categories", fontsize=15)
plt.title("les 10 categories les plus frequent ", fontsize=15)
plt.savefig('10_categorie.png', dpi=400, bbox_inches='tight')
```

**Distribution des catégories**

```python
plt.figure(figsize=(25, 12))
plt.tight_layout()

sns.countplot(Annonces['Categories'])
plt.title("Distribution des catégories", fontsize=15)  
plt.xticks(rotation=70, fontsize=12)
plt.xlabel("categories",fontsize = 20)
plt.ylabel("le nombre de chaque categories",fontsize = 18)
plt.savefig('categorie.png', dpi=400, bbox_inches='tight')
```

```python
plt.figure(figsize=(20, 15))
 # les indices de notre serie (les categories) 
Labels = Serie_categorie.index   
# l'efectif de chaque categorie
Sizes = Serie_categorie         
plt.pie(Sizes, labels=Labels, autopct='%1.1f%%',
        pctdistance = 0.85, labeldistance = 1.2, rotatelabels=False, textprops={'fontsize': 18})
 
plt.show()
```
